```python
import numpy as np 
import matplotlib.pyplot as plt
```

From the notes, the linear convection time marching becomes: $u_{i}^{n+1} = u_{i}^{n} - c \frac{\Delta t}{\Delta x}(u_{i}^{n} - u_{i-1}^n)$

```python
nx = 41
dx = 2/(nx-1)
nt = 25
dt = 0.025
c = 1
```

```python
u = np.ones(nx)
u[int(0.5/dx):int(1/dx)+1] = 2 #Initial conditions of a square wave
plt.plot(u)
```

```python
U = np.zeros((nt,nx))
```

```python
U[0] = u
plt.plot(U[0])
```

```python
for n in range(1, nt):
  for i in range(0, nx):
    U[n,i] = U[n-1,i] - c*dt/dx * (U[n-1,i] - U[n-1,i-1])
```

```python
plt.plot(U[-4])
plt.plot(U[-3])
plt.plot(U[-2])
plt.plot(U[-1])
```

```python
plt.plot(U[0])
plt.plot(U[1])
plt.plot(U[2])
plt.plot(U[3])
```
